# Data Visualization Using Matplotlib & Seaborn Libraries

```python
plt.scatter(data["sepal_length"],data["sepal_width"])
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
```

The above scatter plot gives idea about the distribution and correlation between the entities "sepal_length" and "sepal_width". 

```python
plt.scatter(data["petal_length"],data["petal_width"])
plt.xlabel("petal_length")
plt.ylabel("petal_width")
```

The above scatter plot gives idea about the distribution and correlation between the entities "petal_length" and "petal_width". 

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(2,2,1)
ax.bar(data["species"],data["sepal_length"],data=data)
ax.set_xlabel("species")
ax.set_ylabel("sepal_length")
ax1 = fig.add_subplot(2,2,2)
ax1.bar(data["species"],data["sepal_width"],data=data)
ax1.set_ylabel("sepal_width")
ax1.set_xlabel("species")
ax2 = fig.add_subplot(2,2,3)
ax2.bar(data["species"],data["petal_length"],data=data)
ax2.set_ylabel("petal_length")
ax2.set_xlabel("species")
ax3 = fig.add_subplot(2,2,4)
ax3.set_ylabel("petal_width")
ax3.set_xlabel("species")
ax3.bar(data["species"],data["petal_width"],data=data)
plt.show()
```

```python
data["species"].value_counts().plot.pie(autopct='%1.1f%%')
```

The above pie chart gives an idea about the count number of each species in the dataset. Here, all species are mentioned equal number of times (50 times).

```python
sns.heatmap(data.corr(),annot= True)
```

```python
sns.lineplot(x=data["sepal_length"],y=data["petal_length"],data=data,ci=None,color="magenta")
```

This line plot shows the trends between sepal_length and petal_length. In overall as sepal_length increases, petal_length increases (but slightly decreases at times).

```python
sns.lineplot(x=data["sepal_width"],y=data["petal_width"],data=data,ci=None,color="green")
```

This line plot shows the trends between sepal_width and petal_width. In overall as sepal_width increases, petal_width decreases (but slightly increases at times).

```python
sns.scatterplot(x = "petal_length", y = "sepal_length", data = data)
```

The above scatter plot gives idea about the distribution and correlation between the entities "sepal_length" and "petal_length".
